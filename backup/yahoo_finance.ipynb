```python
import pandas_datareader.data as web
from datetime import datetime
import pandas
import pickle
import json
import os

%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

```python
df = pandas.read_excel(open('set_list.xlsx','rb'))
df.head()
```

```python
start = datetime(1980, 1, 1)
end = datetime.now()
set_list=[]
data={}
sym_key={}
for i in df.iterrows():
    key=i[1]["Symbol"]
    sym="{}{}".format(key,".BK") 
    set_list.append(sym)
    sym_key[key]=i[1]["Company"]
```

```python
json.dump(sym_key, open(os.path.join("data_set","sym_key.json"),"w"),indent=4)
sym_key = json.load( open(os.path.join("data_set","sym_key.json"),"r"))
```

```python
set_list_out=[]
for sym in set_list:
    try:
        df_sym = web.DataReader(sym, 'yahoo', start, end)        
        if df_sym.shape[0]>100:
            print("sym:{} {}".format(sym,df_sym.shape))
            df_sym.to_csv(os.path.join("data_set",sym+".csv"))
            set_list_out.append(sym)
    except:
        pass
set_list_out.sort()
json.dump(set_list_out, open(os.path.join("data_set","set_list.json"),"w"),indent=4)
```
